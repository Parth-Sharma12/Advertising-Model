```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv('advertising.csv')
df.head()
```

```python
df.describe(include='all')
```

```python
sns.pairplot(df,hue='Clicked on Ad')
```

```python
TimeStamped=pd.to_datetime(df['Timestamp'],format='%Y-%m-%d %H:%M:%S')
```

```python
DF=pd.DataFrame(TimeStamped)
```

```python
df['Year']=DF['Timestamp'].dt.year
```

```python
df['Month']=DF['Timestamp'].dt.month
df['Day']=DF['Timestamp'].dt.day
df['Hour']=DF['Timestamp'].dt.hour
df['Minute']=DF['Timestamp'].dt.minute
df['Second']=DF['Timestamp'].dt.second
```

```python
df.drop('Timestamp',inplace=True,axis=1)
```

```python
df
```

```python
DF=df.copy()
```

```python
DF.drop('Country',inplace=True,axis=1)
```

```python
DF.drop('City',inplace=True,axis=1)
DF.drop('Ad Topic Line',inplace=True,axis=1)
```

```python
DF.drop('Year',inplace=True,axis=1)
```

```python
DF
```

```python
sns.boxplot(DF['Age'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
DF.columns
```

```python
X=DF[['Daily Time Spent on Site', 'Age', 'Area Income',
       'Daily Internet Usage', 'Male', 'Month', 'Day', 'Hour',
       'Minute', 'Second']]
```

```python
Y=DF['Clicked on Ad']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.33,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Log_Model=LogisticRegression(max_iter=1000)
Log_Model.fit(x_train,y_train)
prediction=Log_Model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,prediction))
print(confusion_matrix(y_test,prediction))
```
